```python
#Transform the data (Transform non-numerical into numerical col)
from sklearn.preprocessing import LabelEncoder

for column in df_hd_new.columns:
    if df_hd_new[column].dtype == np.number:
        continue
    df_hd_new[column] = LabelEncoder().fit_transform(df_hd_new[column])
```

# Removing Outliers

```python
from scipy.stats import zscore
z=np.abs(zscore(df_hd_new))
z 
```

```python
thershold=3
print(np.where(z<3))
```

```python
df_hd_final = df_hd_new[(z<3).all(axis=1)]
print(df_hd_new.shape, '\t',df_hd_final.shape)
```

# Model Trainning

```python
#saprating into input and output variables

df_x= df_hd_final.drop(columns=['num'])
df_x.head()
```

```python
y=pd.DataFrame(df_hd_final['num'])
y.head()
```

```python
#Scalling the input variable
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x=sc.fit_transform(df_x)
x=pd.DataFrame(x,columns=df_x.columns)
```

```python
#check skewness
x.skew()
```

```python
#Spliting The Data in Train and Target
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=42)
```

```python
print(x_train.shape, '\t', x_test.shape)
```

```python
print(y_train.shape, '\t', y_test.shape)
```

```python
#importing our Models library
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
#import Metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report
```

```python
model=[KNeighborsClassifier(),DecisionTreeClassifier()]
for m in model:
    m.fit(x_train,y_train)
    print('Score of',m,'is:',m.score(x_train,y_train))
    predmodel=m.predict(x_test)
    print('Metrics')
    print('accuracy_score :',accuracy_score(y_test,predmodel))
    print('confusion_matrix :',confusion_matrix(y_test,predmodel))
    print('classification_report:',classification_report(y_test,predmodel))
    print('-----------------------------------------------------------')
    print('\n')
```

```python
svr= SVR(kernel = "linear")
svr.fit(x_train,y_train)
svr.score(x_train,y_train)
print('Score of',svr,'is:',svr.score(x_train,y_train))
predmodel = svr.predict(x_test)

svr= SVR(kernel = "poly")
svr.fit(x_train,y_train)
svr.score(x_train,y_train)
print('Score of',svr,'is:',svr.score(x_train,y_train))
predmodel = svr.predict(x_test)

svr= SVR(kernel = "rbf")
svr.fit(x_train,y_train)
svr.score(x_train,y_train)
print('Score of',svr,'is:',svr.score(x_train,y_train))
predmodel = svr.predict(x_test)
```

```python
#cross validation
from sklearn.model_selection import cross_val_score
kn = KNeighborsClassifier()
scores = cross_val_score(kn,x,y,cv=5).mean()
print(scores)
```

# Using Esemble Technique to Boostup Score

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=20,random_state=42)
rf.fit(x_train,y_train)
print('Score',rf.score(x_train,y_train))
rfpred=rf.predict(x_test)
print('\n')
print('accuracy_score :',accuracy_score(y_test,rfpred))
print('confusion_matrix :',confusion_matrix(y_test,rfpred))
print('classification_report:',classification_report(y_test,rfpred))
```

```python
#cross validation
from sklearn.model_selection import cross_val_score
rf = RandomForestClassifier()
scores = cross_val_score(rf,x,y,cv=5).mean()
print(scores)
```
